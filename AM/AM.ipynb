```python
def find_max_min(lst):
    max_val = lst[0]
    min_val = lst[0]
    
    for i in range(len(lst)):
        if lst[i] > max_val:
            max_val = lst[i]
        if lst[i] < min_val:
            min_val = lst[i]
    
    return max_val, min_val



def count_frequency(lst):
    freq = {}
    
    for i in range(len(lst)):
        element = lst[i]
        
        if element in freq:
            freq[element] = freq[element] + 1
        else:
            freq[element] = 1
    
    return freq
```

Q2

```python
def reverse_number(num):
    rev = 0
    
    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num = num // 10
    
    return rev

def is_palindrome(num):
    original = num
    reversed_num = 0
    
    while num > 0:
        digit = num % 10
        reversed_num = reversed_num * 10 + digit
        num = num // 10
    
    if original == reversed_num:
        return True
    else:
        return False
```

Q3

```python
def tuple_list_to_dict(tuples_list):
    result = {}
    
    for i in range(len(tuples_list)):
        pair = tuples_list[i]
        key = pair[0]
        value = pair[1]
        
        result[key] = value
    
    return result


def max_key_in_dict(d):
    max_key = None
    max_value = None
    
    for key in d:
        if max_value is None or d[key] > max_value:
            max_value = d[key]
            max_key = key
    
    return max_key
```

Q4

```python
def sum_and_average(*args):
    total = 0
    count = 0
    
    for num in args:
        total = total + num
        count = count + 1
    
    if count == 0:
        return 0, 0
    
    average = total / count
    return total, average


def highest_marks(**kwargs):
    top_student = None
    max_marks = None
    
    for name in kwargs:
        if max_marks is None or kwargs[name] > max_marks:
            max_marks = kwargs[name]
            top_student = name
    
    return top_student, max_marks
```

```python
lst = [3, 1, 5, 2, 5, 3, 1]

print("Max and Min:", find_max_min(lst))
print("Frequency:", count_frequency(lst))

num = 121
print("Reverse:", reverse_number(num))
print("Is Palindrome:", is_palindrome(num))

tuple_list = [("a", 10), ("b", 20), ("c", 15)]
print("Converted Dict:", tuple_list_to_dict(tuple_list))

marks_dict = {"A": 50, "B": 80, "C": 70}
print("Max Key:", max_key_in_dict(marks_dict))

print("Sum & Avg:", sum_and_average(10, 20, 30, 40))

print("Top Student:", highest_marks(John=75, Alice=85, Bob=65))
```

```python


class Vector:
    
    def __init__(self, values):
        self.values = values   
    

    def __add__(self, other):
        result = []
        
        for i in range(len(self.values)):
            result.append(self.values[i] + other.values[i])
        
        return Vector(tuple(result))
    
    
    def __sub__(self, other):
        result = []
        
        for i in range(len(self.values)):
            result.append(self.values[i] - other.values[i])
        
        return Vector(tuple(result))
    
    def __mul__(self, scalar):
        result = []
        
        for i in range(len(self.values)):
            result.append(self.values[i] * scalar)
        
        return Vector(tuple(result))
    
    def __repr__(self):
        return "Vector" + str(self.values)

v1 = Vector((1, 2, 3))
v2 = Vector((4, 5, 6))

print("v1:", v1)
print("v2:", v2)
print("Addition:", v1 + v2)
print("Subtraction:", v1 - v2)
print("Scalar Multiplication:", v1 * 3)
```

*args is used to pass multiple values (non-keyword arguments) into a function
It stores them as a tuple
**kwargs is used to pass multiple keyword arguments (key=value)
It stores them as a dictionary

Q3 — Dunder Methods in Python

Dunder methods = Double underscore methods.They are also called magic methods..They get automatically called by Python.
Why important:
They allow us to define custom behavior for objects
Help in operator overloading (like +, -, print)

Examples:
__init__ → constructor (runs when object is created)
__str__ → controls what gets printed
__add__ → used for + operator
